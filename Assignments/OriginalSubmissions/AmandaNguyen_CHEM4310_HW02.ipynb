6. Using the dictionary with the chemical element data.

(a) Copy the dictionary with the chemical element data from Lecture 3 into the colab notebook. Print the value of the atomic mass for oxygen from the dictionary.

```python
Chemical_element_data = {"Hydrogen": {"valence": 1, "atomic_mass": 1.008, "first_ionization_energy": 1310, "electronegativity": 2.1}, "Carbon": {"valence": 4, "atomic_mass": 12.01, "first_ionization_energy": 1090, "electronegativity": 2.5}, "Oxygen": {"valence": 6, "atomic_mass": 16.0, "first_ionization_energy": 1310, "electronegativity": 3.5}}
print(Chemical_element_data["Oxygen"]["atomic_mass"])
```

(b) Add the data for the chemical element fluorine to the dictionary: valence = 7, electronegativity = 4.0, atomic mass = 19.0, first ionization energy = 1680

```python
Chemical_element_data["Fluorine"] = {"valence": 7, "atomic_mass": 19.0, "first_ionization_energy": 1680, "electronegativity": 4.0}
print(Chemical_element_data["Fluorine"])
```

(c) Use the dictionary to check programmatically if carbon or hydrogen has a larger electronegativity or if they are the same size, and print the result.

```python
C = Chemical_element_data["Carbon"]["electronegativity"]
H = Chemical_element_data["Hydrogen"]["electronegativity"]

if C > H:
    statement = "Carbon has a larger electronegativity value than that of Hydrogen."
elif C == H:
    statement = "Carbon has the same electronegativity value as that of Hydrogen."
else:
    statement = "Hydrogen has a larger electronegativity value than that of Carbon."

print(statement)
```

(d) Use the dictionary to check programmatically if oxygen or fluorine has the largest first ionization energy, or if they are the same

```python
O = Chemical_element_data["Oxygen"]["first_ionization_energy"]
F = Chemical_element_data["Fluorine"]["first_ionization_energy"]

if O > F:
    statement = "Oxygen has a greater first ionization energy than Fluorine."
elif O == F:
    statement = "Oxygen has the same first ionization energy as Fluorine."
else:
    statement = "Fluorine has a greater first ionization energy than Oxygen."

print(statement)
```

(e) Use the dictionary to check programmatically if oxygen or carbon has the largest atomic mass, or if they are the same

```python
O = Chemical_element_data["Oxygen"]["atomic_mass"]
C = Chemical_element_data["Carbon"]["atomic_mass"]

if O > C:
    statement = "Oxygen has a greater atomic mass than Carbon."
elif O == C:
    statement = "Oxygen has the same atomic mass as Carbon."
else:
    statement = "Carbon has a greater atomic mass than Oxygen."

print(statement)
```
