1. Exception handling.

(a) Use nested for loops to iterate over all the integers i and j with values from 0 to 10, and calculate idj = i * j / (i - j), catching the error of dividing by 0.

```python
from google.colab import output
output.no_vertical_scroll()

idj = 0

for i in range(0, 10 + 1):
    for j in range(0, 10 + 1):
        try:
            idj = (i * j) / (i - j)
            statement = str(idj)
        except ZeroDivisionError:
            statement = "Cannot divide by zero"

        print("i = " + str(i) + " & j = " + str(j) + ": " + statement)
```

(b) Use nested for loops to iterate through all the values of the list [1.0, 4, "Alice", 21, "42", "Bob"] in both the inner and outer loops, and multiply each value in the list by each other one, catching all of the type errors of multiplying strings.


```python
a_list = [1.0, 4, "Alice", 21, "42", "Bob"]

for i in range(0, len(a_list)):
    for j in range (0, len(a_list)):
        try:
            product = a_list[i] * a_list[j]
            print("i = " + str(i) + " & j = " + str(j) + ": " + str(product))
        except TypeError:
            print("TypeError: Values cannot both be strings.")
```

(a) Copy the text file "Chem_element_data.txt" included with the assignment to the same folder as the notebook in Google Drive (if you are using Colab) or the same directory in your local computer (if using Jupyter). Write a Python code to read and parse this file, and write the included data into a Python dictionary with the same format as the "Chemical_element_data" dictionary from the previous homeworks. You can use the chemical symbols instead of the element names as the dictionary keys. Hint: start with the chem_store function from question 3(a) in Homework 3.

```python
from google.colab import drive
drive.mount('/content/drive')

f = open('/content/drive/MyDrive/Colab Notebooks/Chem_element_data.txt', 'r')
Chemical_element_data = {}

for line in f.readlines():
    if line[-1] == "\n":
        line = line[:-1]  # remove "\n"

    name_list = line.split(":")
    element = name_list[0]
    attributes = name_list[1]
    attributes_and_values = attributes.split(",")
    entry = {}

    for attribute_and_value in attributes_and_values:
        attribute_and_value = attribute_and_value.split("=")
        entry[attribute_and_value[0]] = attribute_and_value[1]

    Chemical_element_data[element] = entry

print(Chemical_element_data)

fout = open('Chemical_element_data.txt', 'w')
fout.write(str(Chemical_element_data))

f.close()
fout.close()
drive.flush_and_unmount()
```

4. (2x) Combining functions and data in classes.

(a) Create a class that combines the class with the storing function and the
extracting function in HW 3, Q6, with an empty dictionary to store the chemical element data. Hint: you will need to use the __init__ method to create the empty dictionary - see slides 14 to 16 of Lecture 6 for an example for lists.

```python
class Chemistry:

    def __init__(self):
        self.data = {}

    def chem_store(self, chem_name, chem_attributes):
      attributes_list = chem_attributes.split(",")
      self.data[chem_name] = {}

      for entry in attributes_list:
          key, value = entry.split("=")
          self.data[chem_name][key] = float(value)

    def chem_lookup(self, element_name, attribute):
        return self.data[element_name][attribute]
```

(b) Create an instance of the class, and load it with the chemical element data dictionary (e.g. from the one loaded in Question 3).

```python
from google.colab import drive
drive.mount('/content/drive')

chemistry_object = Chemistry()

f = open('/content/drive/MyDrive/Colab Notebooks/Chem_element_data.txt', 'r')

for line in f.readlines():
    if line[-1] == "\n":
        line = line[:-1]

    name_list = line.split(":")
    element = name_list[0]
    attributes = name_list[1]
    chemistry_object.chem_store(element, attributes)

f.close()
drive.flush_and_unmount()
print(chemistry_object.data)
```

5. Vectors

(a) Write a function to calculate the scalar product of two vectors, stored in the form of lists.

```python
def scalar_product(x, y):
    if len(x) != len(y):
        raise ValueError("Error: vectors are different lengths.")

    xdoty = 0.0

    for i in range(0, len(x)):
        xdoty += x[i] * y[i]

    return xdoty

G = [1.6, 4.2, 2.4]
H = [3.8, 7.2, 6.5]
print(scalar_product(G, H))
```

(a) Use the numpy module to multiply the following two 3x3 matrices:
\begin{array}{ccc} 3.7\ 7.1\ 6.4 \\
                          5.4\ 6.1\ 2.8 \\
                          4.1\ 2.4\ 8.3  \\ \end{array}
\
\begin{array}{ccc} 6.2\ 5.8\ 4.1 \\
                          3.9\ 4.6\ 6.7 \\
                          9.2\ 5.3\ 4.4  \\ \end{array}


```python
A = np.array([[3.7, 7.1, 6.4], [5.4, 6.1, 2.8], [4.1, 2.4, 8.3]])
B = np.array([[6.2, 5.8, 4.1], [3.9, 4.6, 6.7], [9.2, 5.3, 4.4]])
AB = np.matmul(A, B)
print(AB)
```

(b) Use the numpy module to find the inverse of the matrix that is the result of the multiplication in part (a)

```python
ABinv = np.linalg.inv(AB)
print(ABinv)
```

(c) Use the numpy module to find the eigenvalues and eigenvectors of the matrix resulting from the inversion in part (b)

```python
ABinv_eval, ABinv_evec = np.linalg.eig(ABinv)
print("Eigenvalues: " + str(ABinv_eval))
print("Eigenvectors: " + str(ABinv_evec))
```
