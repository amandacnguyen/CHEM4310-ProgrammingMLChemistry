(a) Write a while loop to print out the squares of all the integers from 1 up to 20

```python
i = 1

while i <= 20:
    print(i ** 2)
    i += 1
```

(b) Copy the while loop above, and modify it to print whether each square is even or odd

```python
i = 1

while i <= 20:
    isq = i ** 2

    if isq % 2 == 0:
        statement = "even"
    else:
        statement = "odd"

    print(str(isq) + " is " + statement + ".")
    i += 1
```

(d) Find the standard deviation of the atomic masses of the elements in the Chemical_element_data dictionary

Standard deviation: stddev.png

```python
meanSum = 0
sdSum = 0

for element in Chemical_element_data:
    meanSum += Chemical_element_data[element]["atomic_mass"]

mean = meanSum / len(Chemical_element_data)

for element in Chemical_element_data:
    sdSum += (Chemical_element_data[element]["atomic_mass"] - mean) ** 2

sd = (sdSum / len(Chemical_element_data)) ** 0.5
print("Standard Deviation: " + str(sd))
```

5. Using functions to perform mathematical algorithms

(a) Implement a function to calculate the factorial of a number. Call this function to calculate the factorials of the integers from 1 to 20.

```python
def factorial(n):
    result = 1

    for i in range(1, n + 1):
        result *= i

    return result


for i in range(1, 20 + 1):
    print(factorial(i))
```

(b) Use the function in (a) to find the lowest value of n where n! > 3**n;
(n! = n factorial = n x n-1 x ... x 3 x 2 x 1)

```python
n = 0

while factorial(n) <= 3 ** n:
    n += 1

print(n)
```

Increment through the integers from 1 to 100
If an integer is divisible by 3, print the string "fizz"
If an integer is divisible by 5, print the string "buzz"
If an integer is divisible by both 3 and 5, print the string "fizzbuzz"

6. (2x) Combining functions in classes

(a) Create a class that combines the storing function in 4(a) and the
extracting function in 4(c), along with functions to calculate the mean and standard deviation of a list of numbers.

```python
class Chemistry:

    def __init__(self):
        self.data = Chemical_element_data

    def chem_store(self, chem_name, chem_attributes):
      attributes_list = chem_attributes.split(",")
      self.data[chem_name] = {}

      for entry in attributes_list:
          key, value = entry.split("=")
          self.data[chem_name][key] = float(value)

    def chem_lookup(self, element_name, attribute):
        return self.data[element_name][attribute]


def findMean(numList):
    sum = 0

    for i in range(len(numList)):
        sum += numList[i]

    return sum / len(numList)



def findSD(numList):
    sum = 0
    mean = findMean(numList)

    for i in range(len(numList)):
        sum += (numList[i] - mean) ** 2

    return (sum / len(numList)) ** (0.5)
```

(b) Use the class to retrieve the required data and calculate and print the average of the first ionization energies for the atoms in the compound NaCl

```python
chemistry_object = Chemistry()
nacl_first_ionization_energy = [
    chemistry_object.chem_lookup("Sodium", "first_ionization_energy"),
    chemistry_object.chem_lookup("Chlorine", "first_ionization_energy"),
]
print(findMean(nacl_first_ionization_energy))
```

(c) Use the class to retrieve the required data and calculate and print the standard deviations of the electronegativities of the atoms in the compound NaCl

```python
nacl_electronegativity = [
    chemistry_object.chem_lookup("Sodium", "electronegativity"),
    chemistry_object.chem_lookup("Chlorine", "electronegativity"),
]
print(findSD(nacl_electronegativity))
```
